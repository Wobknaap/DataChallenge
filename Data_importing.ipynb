```python
import pandas as pd
import json
import os
```

```python
data = []
for root, dirs, files in os.walk('.'):
    for name in files:
        if name.endswith((".json")):
            for line in open(name, 'r'):
                try:
                    add = json.loads(line)
                    data.append(add)
                except:
                    print('errorline')    
                
            
```

```python
df = pd.DataFrame.from_dict(data)
```
