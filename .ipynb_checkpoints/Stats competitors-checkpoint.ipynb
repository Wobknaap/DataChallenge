```python
import math
import datetime
```

```python
time_tweet = df['timestamp_ms'].tolist()
conversat = df['conversation'].tolist()
textlist = df['text'].tolist()


def ResponseTime(mention):
    indexes = []
    convers = []
    conversuniq = []
    conversdup = []
    convtime = []
    response_time = []
    for i in range(len(textlist)):
        if mention in str(textlist[i]):
            indexes.append(i)                           #indexes of tweets mentioning the mention
    for i in range(len(indexes)):
        convers.append(df['conversation'][i])           #conversation numbers of tweets involving the mention
    for val in convers:
        if val in conversuniq:
            conversdup.append(val)
        if val not in conversuniq:
            conversuniq.append(val)                     #list of actual conversations ID's with mention 
    for i in range(len(conversdup)):                    #loop over list of conversations with response.
        index_conv = indexes[i]                         #index in whole dataframe of conversation
        request = time_tweet[index_conv]                #time of mention
        for i in range(len(conversat)):                 #check for conversation value and gather tweets which are equal
            if conversat[i] == conversat[index_conv]:
                convtime.append(time_tweet[i])
        sorted(convtime)
        for i in range(len(convtime)-1):
            if convtime[i] is request:
                response_time.append(convtime[i+1]-convtime[i])
    average_timedelta = sum(response_time, datetime.timedelta(0)) / len(response_time)
    return("The average time it takes for {} to respond to a tweet is {}".format(mention, average_timedelta))
  

ResponseTime("@AmericanAir")
```
