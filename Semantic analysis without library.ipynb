```python
import csv
import sqlite3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk 
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
import re
```

```python
def import_tweets(filename, header = None):
    #import data from csv file via pandas library
    semantic_df = pd.read_csv(filename, encoding = 'latin',header=header)
    #the column names are based on sentiment140 dataset provided on kaggle
    semantic_df.columns = ['sentiment','id','date','flag','user','text']
    #delete 3 columns: flags,id,user, as they are not required for analysis
    for i in ['flag','id','user','date']: 
        del semantic_df[i] # or tweet_dataset = tweet_dataset.drop(["id","user","date","user"], axis = 1)
    #in sentiment140 dataset, positive = 4, negative = 0; So we change positive to 1
    semantic_df.sentiment = semantic_df.sentiment.replace(4,1)
    return semantic_df
```

```python
sentiment_df = import_tweets('semantic2.csv')
```

```python
stopwords_english =  stopwords.words('english')
stemmer_english = SnowballStemmer('english')

text_cleaning_re = "@\S+|https?:\S+|http?:\S|[^A-Za-z0-9]+"

def preprocess(text, stem=False):
    text = re.sub(text_cleaning_re, ' ', str(text).lower()).strip()
    tokens_english = []
    for token in text.split():
        if token not in stopwords_english:
            if stem:
                tokens_english.append(stemmer_english.stem(token))
            else:
                tokens_english.append(token)
    return " ".join(tokens_english)
```

```python
sentiment_df['text'] = sentiment_df['text'].apply(lambda x: preprocess(x))
```
